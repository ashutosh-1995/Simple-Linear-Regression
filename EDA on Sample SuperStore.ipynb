### Now changing all the categorical columns into continuous columns

```python
retail.head()
```

```python
retail['Ship Mode'].unique()
```

```python
retail['Ship Mode'] = retail['Ship Mode'].replace('Standard Class', 0)
retail['Ship Mode'] = retail['Ship Mode'].replace('First Class', 1)
retail['Ship Mode'] = retail['Ship Mode'].replace('Second Class', 2)
retail['Ship Mode'] = retail['Ship Mode'].replace('Same Day', 3)
```

```python
retail['Segment'].unique()
```

```python
retail['Segment'] = retail['Segment'].replace('Consumer', 0)
retail['Segment'] = retail['Segment'].replace('Home Office', 1)
retail['Segment'] = retail['Segment'].replace('Corporate', 2)
```

```python
retail['Region'].unique()
```

```python
retail['Region'] = retail['Region'].replace('East', 0)
retail['Region'] = retail['Region'].replace('West', 1)
retail['Region'] = retail['Region'].replace('South', 2)
retail['Region'] = retail['Region'].replace('Central', 3)
```

```python
retail['Category'].unique()
```

```python
retail['Category'] = retail['Category'].replace('Technology', 0)
retail['Category'] = retail['Category'].replace('Office Supplies', 1)
retail['Category'] = retail['Category'].replace('Furniture', 2)
```

### Checking the correlation of each columns with the target column by creating the heat map

```python
sns.heatmap(retail.corr(), annot= True)
plt.show()
```

 #### By looking at the heat map we can see that there are only two columns that has high correlation with the target column (Profit) which are Sales and Discount. Sales has high Positive correlation with the Profit and Discount has high negative correlation with Profit. Apart from them every other column has very low correaltion with the profit columns. So we can concude from it that if a product has very high sales then the the more profit can be acquired and if the product has vey low or no discount then more profit can be acquired. So we have to increase the sales of the product with giving any discounts on them or very low discounts on them. Hence the task has been completed.
